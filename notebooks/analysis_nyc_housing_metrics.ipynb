```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_parquet("../warehouse/nyc_sales_all.parquet")

# 1. Median sale price by year
plt.figure(figsize=(10,6))
df.groupby("sale_year")["sale_price"].median().plot(
    kind="line",
    title="Median Sale Price by Year (NYC)",
    ylabel="Sale Price ($)"
)
plt.savefig("../visuals/median_sale_price_by_year.png", dpi=300, bbox_inches="tight")
plt.show()
```
